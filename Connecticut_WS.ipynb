```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
df = pd.read_csv('Desktop/Artificial Intelligence/Connecticut Weather_Stations.csv', encoding='unicode_escape')
```

```python
df.head()
```

```python
gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df['Longitude '], df['Latitude ']),
    crs="EPSG:4326"  # Standard WGS84 coordinate reference system
)
gdf.head()
```

```python
# Reads the data
CT = gpd.read_file('Desktop/Artificial Intelligence/Connecticut_Routes_-8583715507025981790/deepgis_DEEP_ROUTES.shp')
CT.crs
```

```python
gdf.crs
```

```python
gdf_projected = gdf.to_crs(CT.crs)
```

```python
gdf_projected.crs
```

```python
gdf_projected.columns
```

```python
# Lengths of variables a, b and c are the same
a = ['Bridgeport Sikorsky Mem Airport', 'Bulls Bridge Dam', 'Burlington', 'Cockaponset Rs', 'Coventry', 'Danbury', 'Falls Village', 'Groton',
     'Hartford Brainard Fd', 'Hartford Bradley International Airport', 'Mansfield Hollow Lake', 'Middletown 4 W', 'Mount Carmel', 'New Haven',
     'Norfolk 2 Sw', 'Norwich Pub Util Plt', 'Shepaug Dam', 'Stamford 5 N', 'Storrs', 'West Thompson Lake', 'Wigwam Reservoir']
b = gdf_projected['Longitude ']
c = gdf_projected['Latitude ']
                
m = folium.Map(location=[c[11], b[11]],
                 zoom_start=9)

# This 'for' loop creates an interactive map with a marker at each of the 21 weather
# stations in Connecticut
for i in range(len(a)):
     folium.Marker([c[i], b[i]],
                  popup=a[i]).add_to(m)  
m
                
```
